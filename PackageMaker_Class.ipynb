```python
import requests as rq
from bs4 import BeautifulSoup as bs
import re
```

```python
class PackageFactory:
    
    def __init__(self,date):
        self.date=date
        return None

    def scraper(self,date):
        date=self.date
        url=f'https://api.composer.nprstations.org/v1/widget/519298c7e1c876ffebb2149b/day?date={date}&format=html'
        html = rq.get(url) 
        soup = bs(html.text, 'html.parser')
        indv_songs=list(soup.find_all('div',class_='single-data'))
        pre_parsed_data=[]
        for x in indv_songs:
            pre_parsed_data.append(x.get_text())
        return pre_parsed_data
        self.list_=pre_parsed_data
    
    def entry_preparer(self,string): 
        title_result=re.search('TITLE:(.*)ARTIST', string)
        if 'COMPOSER' in title_result.group(1):
            composer=re.search('COMPOSER:(.*)ARTIST',string)
            title=re.search('TITLE:(.*)COMPOSER',string).group(1)
        else:
            title=title_result.group(1)
        artist_result=re.search('ARTIST:(.*)ALBUM',string)
        artist_result_0=re.search('ARTIST:(.*)',string)
        if bool(artist_result)==True:
            artist=artist_result.group(1)
        elif bool(artist_result_0)==True:
            artist=artist_result_0.group(1)
        else:
            artist='Blank'
        album_result=re.search('ALBUM:(.*)',string)
        if bool(album_result)==True:
            album=album_result.group(1)
        else:
            album='Blank'
        entry=(title,artist,album)
        return entry
    
    def package_maker(self):
        list_=self.scraper(self.date)
        package=[]
        for item in list_:
            single=self.entry_preparer(item)
            package.append(single)
        return package
```
